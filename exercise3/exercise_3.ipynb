```python
#Q1 a)
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread("butterfly.jpg")
imgrgb = cv.cvtColor(img , cv.COLOR_BGR2RGB)

#box filtering
kernel = np.ones((9,9), np.float32)/81
imgcA = cv.filter2D(img, -1, kernel)

# apply a "Gaussian" blur to the image
imgcB = cv.GaussianBlur(img, (9, 9), 4)

imgcA = cv.cvtColor(imgcA, cv.COLOR_BGR2RGB)
imgcB = cv.cvtColor(imgcB, cv.COLOR_BGR2RGB)

fig,axes =plt.subplots(2 ,2 ,figsize = (18,10) )
plt.rcParams["figure.autolayout"] = True

axes[0][0].imshow(imgrgb)
axes[0][0].set_title('Original')

axes[0][1].imshow(imgcA)
axes[0][1].set_title('Averaging')

axes[1][0].imshow(imgrgb)
axes[1][0].set_title('Original')

axes[1][1].imshow(imgcB)
axes[1][1].set_title('Gaussian')

axes[0][0].set_xticks([]), axes[0][0].set_yticks([])
axes[1][0].set_xticks([]), axes[1][0].set_yticks([])
axes[0][1].set_xticks([]), axes[0][1].set_yticks([])
axes[1][1].set_xticks([]), axes[1][1].set_yticks([])
plt.show()


```

```python
#Q3)
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread("contact_lens.tif" , cv.IMREAD_REDUCED_GRAYSCALE_2)

# Sobel vertical
kernel = np.array([(-1, -2, -1), (0, 0, 0), (1, 2, 1)], dtype='float')
imgc = cv.filter2D(img,-1,kernel)


fig,axes =plt.subplots(1 ,2 ,figsize = (18,18) )

axes[0].imshow(img , cmap=  'gray')
axes[0].set_title('Original')

axes[1].imshow(imgc , cmap=  'gray' , vmin = -1020 , vmax = 1020)
axes[1].set_title('Sobel Vertical')

axes[0].set_xticks([]), axes[0].set_yticks([])
axes[1].set_xticks([]), axes[1].set_yticks([])

plt.show()
```
